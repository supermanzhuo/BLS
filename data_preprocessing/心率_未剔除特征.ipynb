```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', None)#使其显示完全
```

```python
#获取数据集

data=pd.read_csv(r"C:\Users\ZHUO\Desktop\start_to_stop.csv")
data.shape #(4547,185) 原来是189列 后去掉时间三列和event_id
```

```python
data.info()
data.head()
```

```python
#计算每一列NaN的占比
data.isnull().sum().sort_values(ascending=False)/data.shape[0]
```

```python
data.isnull().sum()/data.shape[0]
```

```python
#统计数据集中有哪些列包含NaN
data_nan=[]

for col in data.columns:
    if data[col].isnull().sum()>0:
        data_nan.append(col)
        
len(data_nan)
```

```python
#针对空缺值的处理方式
#折中法：对于数值列类型 取中位数；对于分类列，取NaN

for col in data_nan:
    data[col].fillna(data[col].median(),inplace=True)#中位数代替

data.shape
```

```python
data.to_csv('dataset.csv')
```

```python
#方差过滤
from sklearn.feature_selection import VarianceThreshold
X=data
selector=VarianceThreshold()
X_fsvar=selector.fit_transform(X)
X_fsvar.shape
```

```python
X['altitude_change_past_10800_to_14400_seconds'].var()
```

```python
#新增一列作为标签
data['heart_rate_change_value']=data["heart_rate_stop"]-data["heart_rate_start"]
data.shape
```

```python
data.head()
```

```python
#归一化
from sklearn.preprocessing import MinMaxScaler
x=data.drop(columns=['heart_rate_change_value'],axis=1).values
scaler = MinMaxScaler()  #实例化
X= scaler.fit_transform(x)
print(X)

t=data['heart_rate_change_value'].values.reshape(-1,1) #转换成一列
y=scaler.fit_transform(t)
print(y)
```

```python
#数据集分离
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
#将X_train和y_train合并
xt=pd.DataFrame(X_train)
yt=pd.DataFrame(y_train)
xt.columns=data.columns[:-1]
yt.columns=['heart_rate_change_value']
xyt=pd.concat((xt,yt),axis=1)
xyt.head()
xyt.shape

xyt.to_csv('train.csv')  #3182
```

```python
xt=pd.DataFrame(X_train)
xt.head()
```

```python
#将X_test和y_test合并
xe=pd.DataFrame(X_test)
ye=pd.DataFrame(y_test)
xe.columns=data.columns[:-1]
ye.columns=['heart_rate_change_value']
xye=pd.concat((xe,ye),axis=1)
xye.head()
xye.shape

xye.to_csv('test.csv')
```

```python
#未经归一化后的数据集分离
from sklearn.model_selection import train_test_split
X_train1,X_test1,y_train1,y_test1=train_test_split(x,t,test_size=0.3,random_state=10)
```

```python
#将X_train1和y_train1合并
xt1=pd.DataFrame(X_train1)
yt1=pd.DataFrame(y_train1)
xt1.columns=data.columns[:-1]
yt1.columns=['heart_rate_change_value']

xyt1=pd.concat((xt1,yt1),axis=1)
xyt1.shape

xyt1.to_csv('train1.csv')
```
